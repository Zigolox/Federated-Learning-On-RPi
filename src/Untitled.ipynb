```python
from runNClients import testNClients
import matplotlib.pyplot as plt
```

```python
clients = await testNClients(2,2,10)
```

```python
def plot_data(clients):
    def plot(client_data,idx):
        for i,data in enumerate(client_data):
            plt.plot(list(range(1,len(data[idx])+1)),data[idx],label=f"client {i}")
            plt.xticks(list(range(1,len(data[idx])+1)))
        plt.legend()
        plt.show()


    plt.title("Test accuracy")
    plt.xlabel("Number of Rounds")
    plt.ylabel("Accuracy")
    plot(clients,0)
    plt.title("Test loss")
    plt.xlabel("Number of Rounds")
    plt.ylabel("Loss")
    plot(clients,1)
    plt.title("Train loss")
    plt.xlabel("Total number of epochs")
    plt.ylabel("Loss")
    plot(clients,2)

plot_data(clients)
```
